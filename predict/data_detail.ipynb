```python
total_num, total_type = read_data.get_data_detail()
```

```python
from prettytable import PrettyTable
table1 = PrettyTable(['编号', '目录/日期', '文件数'])
all_datafile_sum = 0
table1_id = 0
for key, value in total_num.items():
    table1.add_row([table1_id, key, value])
    all_datafile_sum += value
    table1_id += 1
table1.add_row(['Total','',all_datafile_sum])
print(table1)
```

```python
table2 = PrettyTable(['类型', '目录', '文件数'])
for key, value in total_type.items():
    total = 0
    for d, f_l in value.items():
        table2.add_row([key, d, len(f_l)])
        total += len(f_l)
    table2.add_row(['Total', '', total])
print(table2)
```
